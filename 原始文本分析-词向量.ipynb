```python
source = []
target = []
weight = []
for b1, b2 in bb.keys():
    source.append(b1)
    target.append(b2)
    w = bb[(b1, b2)]
    if b1 in adjust_dict:
        w = w * adjust_dict[b1]
    if b2 in adjust_dict:
        w = w * adjust_dict[b2]
    weight.append(w)
```

```python
edges = [(source[i], target[i], {'weight': weight[i]}) for i in range(len(source))]
```

```python
import networkx as nx
G = nx.Graph()
```

```python
G.add_nodes_from(source)
```

```python
G.add_edges_from(edges)
```

```python
pg = nx.pagerank(G)
```

```python
model = gensim.models.KeyedVectors.load('second.model')
```

```python
for b in brand_list:
    if b not in model:
        print(b)
```

```python
n_brand_list = [b for b in brand_list if b in model.wv]
```

```python
n_word_list = [w for w in word_list if w in model.wv]
```

```python
score = {}
for w in n_word_list:
    w_score = [model.wv.similarity(b, w) for b in n_brand_list]
    score[w] = w_score
```

```python
score_df = pd.DataFrame(score, index=n_brand_list)
```

```python
score_df['雷同'] = -score_df['雷同']
score_df['偷税漏税'] = -score_df['偷税漏税']
score_df['污染'] = -score_df['污染']
score_df['下岗率'] = -score_df['下岗率']
```
